# The Shopping Cart Program

```python
from tabulate import tabulate
print("                      STOCKS     ")
data=[["s.no","item","Quantity","Cost/Item"],
    [1,"Biscuts",5,20.50],
     [2,"cereals",5,90],
     [3,"chicken",20,200],
     [4,"Oats",20,39.00],
     [5,"rice",30,29.99]]
print(tabulate(data,headers="firstrow",tablefmt="fancy_grid"))
bis=0
cer=0
chi=0
oats=0
rice=0
Name=0
Address=0
Distance=0
while(True):
    print("WELCOME TO SHOPPING MART")
    while(True):
        bis=int(input("How many Biscuits are required:-"))
        if data[1][2]==0:
            print("out of stock")
            bis=0
            break
        elif bis<=data[1][2]:
                break
        else:
                print("only",bis,"biscuits are available")
    
    while(True):
        cer=int(input("How many cereal packets are reqired:-"))
        if data[2][2]==0:
            print("out of stock")
            cer=0
            break
        elif cer<=data[2][2]:
            break
        else:
            print("only",cer,"cereals are available")
        
    
    while(True):
        chi=int(input("How many Frozen chicken packets are reqired:-"))
        if data[3][2]==0:
            print("out of stock")
            chi=0
            break
        elif chi<=data[3][2]:
            break
        else:
            print("only",chi,"chicken are available")
        
    
    while(True):
        oats=int(input("How many oats packets are reqired:-"))
        if data[4][2]==0:
            print("out of stock")
            oats=0
            break
        elif oats<=data[3][2]:
            break
        else:
            print("only",oats,"oats are available")
        
    
    while(True):
        rice=int(input("How many rice packets are reqired:-"))
        if data[5][2]==0:
            print("out of stock")
            rice=0
            break
        elif rice<=data[3][2]:
            break
        else:
            print("only",rice,"rice are available")
    print()
    print("Fill in your delivery details")
    name=input("Enter your name:")
    Address=input("Enter your Address:")
    Distance=int(input("enter your distance"))
    delivery_charge = 2 * Distance
    print("delivery charge:",delivery_charge,"will be leived for",Distance,"km of distance")
    print("Sucessfully completed ....")
    break
  
bill=[["s.no","item","Quantity","TotalCost"],
    [1,"Biscuts",bis,bis*20.50],
     [2,"cereals",cer,cer*90],
     [3,"chicken",chi,chi*200],
     [4,"Oats",oats,oats*39.00],
     [5,"rice",rice,rice*29.99]]
###reduse data using bill list
data[1][2] -= bis
data[2][2] -= cer
data[3][2] -= chi
data[4][2] -= oats
data[5][2] -= rice
# Calculate the total cost for each item in the bill
total_cost_bis = bis * 20.50
total_cost_cer = cer * 90
total_cost_chi = chi * 200
total_cost_oats = oats * 39.00
total_cost_rice = rice * 29.99
total_bill = total_cost_bis + total_cost_cer + total_cost_chi + total_cost_oats + total_cost_rice
back=input("Do you want to Continue Shopping? YES OR NO")
if back.lower()=="yes":
    bis1=0
    cer1=0
    chi1=0
    oats1=0
    rice1=0
    while(True):
        while(True):
            bis1=int(input("How many Biscuits are required:-"))
            if data[1][2]==0:
                print("out of stock")
                bis1=0
                break
            elif bis1<=data[1][2]:
                break
            else:
                print("only",data[1][2],"biscuits are available")

        
        while(True):
            cer1=int(input("How many cereal packets are reqired:-"))
            if data[2][2]==0:
                print("out of stock")
                cer1=0
                break
            elif cer1<=data[2][2]:
                break
            else:
                print("only",data[2][2],"cereals are available")
        
        while(True):
            chi1=int(input("How many Frozen chicken packets are reqired:-"))
            if data[3][2]==0:
                print("out of stock")
                chi1=0
                break
            elif chi1<=data[3][2]:
                break
            else:
                print("only",data[3][2],"chicken are available")
        
        while(True):
            oats1=int(input("How many oats packets are reqired:-"))
            if data[4][2]==0:
                print("out of stock")
                oats1=0
                break
            elif oats1<=data[4][2]:
                break
            else:
               print("only",data[4][2],"oats are available")
        
        while(True):
            rice1=int(input("How many rice packets are reqired:-"))
            if data[5][2]==0:
                print("out of stock")
                rice1=0
                break
            elif rice1<=data[5][2]:
                break
            else:
               print("only",data[5][2],"rice are available")
        break
    bill=[["s.no","item","Quantity","TotalCost"],
    [1,"Biscuts",bis+bis1,(bis+bis1)*20.50],
    [2,"cereals",cer+cer1,(cer+cer1)*90],
    [3,"chicken",chi+chi1,(chi+chi1)*200],
    [4,"Oats",oats+oats1,(oats+oats1)*39.00],
    [5,"rice",rice+rice1,(rice+rice1)*29.99]]
    print("Move to Your Card Details:")
    print()
    print("                         BILL     ") 
    print(tabulate(bill,headers="firstrow",tablefmt="fancy_grid"))
    data[1][2] -= bis1
    data[2][2] -= cer1
    data[3][2] -= chi1
    data[4][2] -= oats1
    data[5][2] -= rice1

    total_cost_bis1 = bis1 * 20.50
    total_cost_cer1 = cer1 * 90
    total_cost_chi1 = chi1 * 200
    total_cost_oats1 = oats1 * 39.00
    total_cost_rice1 = rice1 * 29.99
    total_bill1 = total_cost_bis1 + total_cost_cer1 + total_cost_chi1 + total_cost_oats1 + total_cost_rice1
    Final_bill=total_bill1+total_bill
    print("Total Item Cost:",Final_bill)
    print()
    print("Package Address:")
    print("Name:",name)
    print("Address:",Address)
    print()
    print("Total Bill Amount: Total Item Cost+delivery charge :",Final_bill+delivery_charge)
else:
    print("Move to Your Card Details:")
    print()
    print("                       BILL     ") 
    print(tabulate(bill,headers="firstrow",tablefmt="fancy_grid"))
    print("Total Item Cost:",total_bill)
    print()
    print("Package Address:")
    print("Name:",name)
    print("Address:",Address)
    print()
    print("Total Bill Amount: Total Item Cost+delivery charge :",total_bill+delivery_charge)
print("Have a Nice Day")
    
```
