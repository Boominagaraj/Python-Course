# Task-1)

## Assume that you have some data of customers with Cust_ID in one sheet of an excel file and some more data with Cust_ID in a second sheet of the same excel file. How can you join these data?
### Read data from first sheet of excel file 'customer.xlsx'

```python
import pandas as pd
```

```python
df = pd.read_excel("/content/DS1_C4_S4_Customer_Data_Concept.xlsx")
```

```python
df1 = pd.read_excel("/content/Customer_join_data.xlsx")
```

```python
df
```

```python
df1
```

### Read the data from second sheet of excel file.

## 1. Inner Join

```python
pd.merge(df, df1)
```

```python
df.set_index('Cust_ID').join(df1.set_index('Cust_ID'), on = 'Cust_ID', how = 'inner')
```

## 2. Outer Join

```python
df.set_index('Cust_ID').join(df1.set_index('Cust_ID'), on = 'Cust_ID', how = 'outer')
```

```python
pd.merge(df, df1, how='outer')
```

## 3. Right Join

```python
pd.merge(df, df1, how = 'right')
```

## 4. Left Join

```python
pd.merge(df, df1, how='left')
```

# Task-17)

## Find out the total missing values in each column of the dataset. Then replace the missing values by appropriate values.
### Total missing values in each column

```python
df4.isnull().sum()
```

### Count the different values of diabetic column

```python
df4.diabetic.value_counts()
```

```python
df4.diabetic.unique()
```

### Replace the missing value by the most frequent value

```python
df4.diabetic.mode()[0]
```

```python
df4.diabetic = df4.diabetic.fillna(df4.diabetic.mode()[0])
```

```python
df4.isnull().sum()
```

### Replace the missing value by mean value in 'claim'

```python
df4.claim = df4.claim.fillna(df4.claim.mean())
```

### Find the total missing values in each column again

### There are no missing values left.

```python
df4.isnull().sum()
```
