# Scenario-1)

### Flattom maintains an active lifestyle and shares his activities on social media. He maintains a daily log with activity details. He has calculated the average number of days spent on different activities in a month based on daily log.

### Flattom wants to share the information on social media, but he is not keen on sharing the data as is. So, he decides to prepare visuals.

Refer DS1_C4_S5_Activity_Lifestyle_Data_Concept.csv file for this task.


## Can you help Flattom to represent the activities of the month on the social media?

```python
df=pd.read_csv("DS1_C4_S5_Activity_Lifestyle_Data_Concept.csv")
df
# Trend alwasys go to line chart
```

```python
df.info()
```

```python
df[['Activity']]
```

```python
plt.plot(df.Activity,df['Average days in a month'],label='Activity',color='green')
plt.legend()
plt.show()
```

```python
plt.plot(df.Activity,df['Average days in a month'],color="grey",label="Trend of Acitivies");
plt.title("Activities of the month on the social media")
plt.legend()
plt.show()
```

## Let's draw Pie Chart

```python
plt.pie(df['Average days in a month'],labels=df.Activity,autopct="%0.2f%%",explode=[0.1,0,0,0,0],shadow=False);
plt.title("Activity Chart");
```

## Line chart vs Pie chart

```python
plt.figure(figsize=(15,8))

plt.subplot(1,2,1)
plt.plot(df.Activity,df["Average days in a month"],color="y",label="Trend of Acitivies")
plt.title("Activities of the month on the social media")
plt.legend()

plt.subplot(1,2,2)
plt.pie(df['Average days in a month'],labels=df.Activity,autopct="%0.2f%%",explode=[0.1,0,0,0,0],shadow=True);
plt.title("Activity Chart");

plt.show()
```

### It is difficult to interpret the proportion of any activity from Line Chart. Pie Chart is recommended to present the composition of data with proportion.

## Use Pie Charts to visually show the composition or 'part of the whole' or the contribution of each category.

# Scenario-3)

## The following data shows brand-wise revenue details for a Sports Accessories Selling store over three consecutive years.

## How can you present brand- and year-wise total revenue from this data? Can you identify the brand that earned the highest revenue over the 3 years?

Refer DS1_C4_S5_Revenue_Data_Concept.csv file for this task.


## First, read the data from the csv file.

```python
df3=pd.read_csv("DS1_C4_S5_Revenue_Data_Concept.csv ")
df3
```

```python
mean=df3.groupby('Brand Name').mean()
mean.plot(kind='bar',color=['red','yellow','black'],xlabel='Brand names',ylabel='mean value',stacked=True)
```

## Create Grouped Bar chart (Clustered Bar Chart)
### Method 1

```python
df3.plot.bar(x="Brand Name",figsize=(12,5))
plt.show()
```

### Method 2

```python
plt.figure(figsize=(12,5))
y=df3["2018"]
y1=df3["2019"]
y2=df3["2020"]
x=df3["Brand Name"]

plt.bar(x,y,color="c")
plt.bar(x,y1,bottom=y)
plt.bar(x,y2,color="orange",bottom=y+y1)

plt.show()
```

## Can you quickly interpret the brand that provides highest net revenue over the period of 3 years?

## Create a Stacked Bar Chart

## Can you quickly interpret the brand that provides the highest net revenue over 3 years?
## YES

### Grouped Bar Chart (clustered bar chart): It is difficult to interpret by comparing the lengths of the rectangular bars of different categories.
### Stacked Chart: It is easy to compare the total segments of various categories on the basis of their bar lengths.
