Observation: There is huge gap between population mean and sample mean, so this sample does not represent properly the population. Do we have any specific technique to extract sample? 

```python
def popul(col):
    print("Mean for the ",col,"is",df[col].mean())
    print("Median for the ",col,"is",df[col].median())
    print("Mode for the ",col,"is",df[col].mode()[0])
    print("Standard Deviation for the ",col,"is",st.stdev(df[col]))
```

```python
popul("Annual_Salary($)")
```

```python
annual_data=sorted(df["Annual_Salary($)"])
print("Kurtosis for annual Salary",kurtosis(annual_data))
```

```python
pop_skew=df["Annual_Salary($)"].skew()
print(pop_skew)
```

## Demo - 3.2: Simple Random Sampling

### Simple Random Sampling without Replacement

```python
random_sample=df.sample(n=20,replace=False,random_state=42)
random_sample
```

### Simple Random Sampling with Replacement

```python
random_sample=df.sample(n=20,replace=True,random_state=42)
random_sample
```

### Random Sampling with a fraction of data from Population

```python
random_sample_40=df.sample(frac=0.4,replace=True,random_state=42)
random_sample_40
```

## Demo - 3.4: Stratified Sampling

```python
g_d=df.groupby("Department")
eng=df.Department.unique()
strat_samp=pd.DataFrame()
for i in eng:
    s_d=g_d.get_group(i)
    s_d_1=s_d.sample(n=5,random_state=1)
    strat_samp=pd.concat([strat_samp,s_d_1],ignore_index=True)
strat_samp
```

## Demo - 3.5: Cluster Sampling

```python
c_s=["IT"]
cluster_sampling=pd.DataFrame()
for i in eng:
    s_d1=g_d.get_group(i)
    s_d2=s_d1.sample(frac=0.3,random_state=1)
    cluster_sampling=pd.concat([cluster_sampling,s_d2],ignore_index=True)
cluster_sampling
```
