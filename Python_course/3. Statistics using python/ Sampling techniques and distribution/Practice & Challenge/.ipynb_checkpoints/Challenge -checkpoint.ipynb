### 3.1.4 Stratified sampling with respect to Year

```python
g_d=df.groupby('Year')
eng=df.Year.unique()
strat_sam=pd.DataFrame()

for i in eng:
    s_d=g_d.get_group(i)
    s_d_1=s_d.sample(n=5,random_state=1)
    strat_sam=pd.concat([strat_sam,s_d_1],ignore_index=True)

strat_sam
```

### 3.1.4 Stratified sampling with respect to Genre

```python
g_d=df.groupby('Genre')
eng=df.Genre.unique()
strat_sam=pd.DataFrame()

for i in eng:
    s_d=g_d.get_group(i)
    s_d_1=s_d.sample(n=5,random_state=1,replace=True)
    strat_sam=pd.concat([strat_sam,s_d_1],ignore_index=True)

strat_sam
```

### 3.1.4- Clustered sampling with respect to year

```python
c_s=df['Year']
cluster_sampling=pd.DataFrame()

for i in c_s:
    s_d1=g_d.get_group(i)
    s_d2=s_d1.sample(frac=0.3,random_state=1)
    cluster_sampling=pd.concat([cluster_sampling,s_d2],ignore_index=True)

cluster_sampling
```

```python
c_s=df['Genre']
cluster_sampling=pd.DataFrame()

for i in c_s:
    s_d1=g_d.get_group(i)
    s_d2=s_d1.sample(frac=0.3,random_state=1)
    cluster_sampling=pd.concat([cluster_sampling,s_d2],ignore_index=True)

cluster_sampling
```

### 3.2.1- Measure the mean,Median and Standard Deviation

```python
sample_10 = df.sample(n = 10, replace = True, random_state = 42)
sample_20 = df.sample(n = 20, replace = True, random_state = 42)
sample_30 = df.sample(n = 30, replace = True, random_state = 42)
sample_40 = df.sample(n = 30, replace = True, random_state = 42)
```

### 10-Samples

```python
def info_cat(col):
    print('Mean for the 10 Sample is', sample_10[col].mean())
    print('Median for the 10 Sample is', sample_10[col].median())
    print('Mode for the 10 Sample is',sample_10[col].mode()[0])
    print('Standard Deviation for the 10 Sample is', st.stdev(sample_10[col]))
```

```python
info_cat('Global_Sales_M$')
```
