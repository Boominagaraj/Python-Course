# Analysis of Muscle Cars

### 3.1.1- Random sampling without replacement


```python
random_sample = df.sample(n = 20, replace = False, random_state = 42)
random_sample
```

### 3.1.2-Random sampling with replacement

```python
random_sample = df.sample(n = 20, replace = True, random_state = 42)
random_sample
```

### 3.1.3 Systematic sampling

```python
index = np.arange(0, len(df),2)
df.iloc[index]
```

### 3.1.4 Stratified sampling with respect to engine type

```python
g_d=df.groupby('vs')
eng=df.vs.unique()
strat_sam=pd.DataFrame()

for i in eng:
    s_d=g_d.get_group(i)
    s_d_1=s_d.sample(n=5,random_state=1)
    strat_sam=pd.concat([strat_sam,s_d_1],ignore_index=True)

strat_sam
```

### 3.1.4- Clustered sampling with respect to engine

```python
c_s=df['vs']
cluster_sampling=pd.DataFrame()

for i in c_s:
    s_d1=g_d.get_group(i)
    s_d2=s_d1.sample(frac=0.1,random_state=1)
    cluster_sampling=pd.concat([cluster_sampling,s_d2],ignore_index=True)

cluster_sampling
```

### 3.2.1- Measure than mean .median and Standard Deviation

```python
sample_10 = df.sample(n = 10, replace = True, random_state = 42)
sample_20 = df.sample(n = 20, replace = True, random_state = 42)
sample_30 = df.sample(n = 30, replace = True, random_state = 42)
```

### Sample-10

```python
def info_cat(col):
    print('Mean for the 10 Sample is', sample_10[col].mean())
    print('Median for the 10 Sample is', sample_10[col].median())
    print('Mode for the 10 Sample is',sample_10[col].mode()[0])
    print('Standard Deviation for the 10 Sample is', st.stdev(sample_10[col]))
```

```python
info_cat("hp")
```

### Distribution

```python
sb.distplot(sample_10['hp'])
```
