### Task-1

```python
import scipy.stats as sts
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sb
import math
```

### The Fewer than five purchasers choose oreo

```python
p=0.15
n=25
k=np.arange(0,5)
prob = sts.binom.pmf(k, n, p)
prob
print("The Probablity of Five purchasers :",sum(prob))
```

### More than 10 purchasers choosing oreo

```python
p=0.15
n=25
k1=np.arange(11,26)
prob1 = sts.binom.pmf(k1, n, p)
prob1
print("The Probablity of Ten purchasers :",sum(prob1))
```

### c)-Mean,Standard Deviation

```python
mean = n*p
std = math.sqrt(n*p*(1-p))
print("Mean of the Purchasers",mean)
print("Stantard Deviation of Purchasers",std)
```

```python
sb.lineplot(x=k,y=prob,label="Five Purchasers")
sb.lineplot(x=k1,y=prob1,label="Ten Purchasers")
plt.xlabel("no of outcome of purchasers")
plt.ylabel("Probablity of eash purchasers")
plt.title("Probablity of Binominal Distribution")
plt.show()
```

### Task-2

##### the data was collected for the total of 60 minutes 6pm-7pm(Every 5 minutes-122 interval)each
##### Saturday night the total number of interval observed would be 36 (3 weeks - 12 intervals/weeks)                      

```python
week1=(3+6+4+6+2+3+1+5+1+0+3+3)/12
week2=(1+2+4+0+2+6+5+4+2+5+3+4)/12
week3=(5+3+5+3+5+4+7+3+4+8+3+1)/12
average_rate=(week1+week2+week3)/3
average_rate
```

### A)-No customers arrive in 5 minutes

```python
rate1=average_rate
n1=0

prob=sts.poisson.pmf(n1,rate1)
prob
print("The Probablity of no customers in 5 minutes :",sum(prob1))
```

### B)- Six or more customers in 5 minutes

```python
rate2=average_rate
n2=np.arange(6,9)

prob2=sts.poisson.pmf(n2,rate2)
print("The Probablity of six or more customers in 5 minutes :",sum(prob2))
```

### C)- Four customers in 10 minutes 

```python
rate3=average_rate*2
n3=np.arange(0,4)

prob3=sts.poisson.pmf(n3,rate3)
print("The Probablity of Four customers in 10 minutes :",sum(prob3))
```

### d)-between  three and six in 10 minutes

```python
rate4=average_rate*2
n4=np.arange(3,7)

prob4=sts.poisson.pmf(n4,rate4)
print("The Probablity of Three to six customers in 10 minutes :",sum(prob4))
```

### E)-Eight customers in 15 minutes

```python
rate5=average_rate*3
n5=np.arange(0,9)

prob5=sts.poisson.pmf(n5,rate5)
print("The Probablity of eight customers in 15 minutes :",sum(prob5))
```

### Distribution

```python
sb.lineplot(x=n2,y=prob2,label=">6-Customer")
sb.lineplot(x=n3,y=prob3,label="4-Customer")
sb.lineplot(x=n4,y=prob4,label="3 to 6-Customer")
sb.lineplot(x=n5,y=prob5,label="8-Customer")
plt.xlabel("no of customer of minutes")
plt.ylabel("Probablity of eash purchasers")
plt.title("Probablity of Poisson Distribution")
plt.show()
```

### Task-1

#### To find the no of construction can't work during summer

```python
X=[6,7,8,9,10,11,12,13,14]
P_x=[0.03,0.08,0.15,0.20,0.19,0.16,0.10,0.07,0.02]
```

```python
plt.figure(figsize=(10,10))
sb.barplot(x=X, y=P_x)
plt.xlabel('X:  no of Days')
plt.ylabel('P(X): Probability of woring in summer')
plt.title('Probability distribution of working in summer')
plt.xticks(rotation=90)
for i in range(len(X)):
    plt.text(i, round(P_x[i],2), str(round(P_x[i],2)), ha = 'center', va = 'bottom')

plt.show()
```

### A)-Does not exceed 9 days the next summer

```python
su=0.03+0.08+0.15
su
```

### b)-7 to 13 days next summer

```python
su1=0.08+0.15+0.2+0.19+0.16+0.1+0.07
su1
```

#### D)-Mean & Standard Deviation

```python
import math
mean=0
var=0
std=0

for x,p_x in zip(X,P_x):
    mean+=x*p_x

for x,p_x in zip(X,P_x):
    var+=(x-mean)**2*p_x

std=math.sqrt(var)
print("Mean :",round(mean,2))
print("Standard Deviation :",round(std,2))
```
