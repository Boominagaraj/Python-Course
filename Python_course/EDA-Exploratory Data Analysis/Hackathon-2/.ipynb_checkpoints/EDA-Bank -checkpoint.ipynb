```python
df["OCCUPATION_TYPE"].fillna(df["OCCUPATION_TYPE"].mode()[0],inplace=True)
```

```python
df["OCCUPATION_TYPE"].isnull().sum()
```

```python
df["CNT_FAM_MEMBERS"].fillna(df["CNT_FAM_MEMBERS"].mode()[0],inplace=True)
```

```python
def info_cat(col):
    print(f"The Unique values for the column {col}are {df[col].unique()}")
    print(f" Most Frequent value for the column {col} is {df[col].mode()[0]}")
    print(f"Total missing value present in the column{col}is {df[col].isnull().sum()}")
```

```python
info_cat("NAME_INCOME_TYPE")
```

```python
def info_num(num):
    print(f"The mean value of the column {num} is {df[num].mean()}")
    print(f"The Median value of the column {num} is {df[num].median()}")
    print(f"The Standard Deviation value for the column {num}is {df[num].isnull().sum()}")
```

```python
info_num("AMT_INCOME_TOTAL")
```

# Family Wise count percentage

```python
# Plot the graph-
fig, ax = plt.subplots(figsize= (10, 7))
ax = sns.countplot(x = df["NAME_FAMILY_STATUS"])
ax.set_xticklabels(ax.get_xticklabels(), rotation = 90, ha = "right") # set_xticklabels is used to set the labels of the x axis
                                                                       # ax.get_xtickslabels() returns all the labels that we need to print
                                                                       # rotation: rotate the xlabels at angle provided, and
                                                                       # ha: Horizontal axis:-> how you want to keep it. can be changed to center, left, right
count = len(df["NAME_FAMILY_STATUS"])                                     # counting the number of rows in whole dataset
for bar in ax.patches:                                                 # ax.patches returns the rectange object from matplotlib 
    percentage = f"{round(bar.get_height()/count, 2)*100} %"           # first it calculate the bar heightand then dividing by the number of row(to calculate percentage)
                                                                            # in data and multiply with 100 to get the percentage and rounding of by 2 numbers.  
    x = bar.get_x() + bar.get_width() /2                                # get_x: returns the left corrdinate of the rectange, get_width: returns width of rectagle
    y = bar.get_height()                                                # get_height: returns height of rectangle bar.
    ax.annotate(percentage, (x, y), ha = "center", va = "top")          # va: Vertical axis: -> top means percentage would bneath the bar head
                                                                        # annotate: used to annotate over objects. here objects are the bars 
plt.show()
```
