```python
def info_cat(col):
    print(f'The unique values for the column {col} are {df[col].unique()}')
    print('Mode for the', col, 'is', df[col].mode()[0])
    print('Total count of missing values in', col, 'is', df[col].isnull().sum())
```

```python
def info_con(col):
    print('The mean of', col, 'is', df[col].mean())
    print('The mode of', col, 'is', df[col].mode()[0])
    print('The median of', col, 'is', df[col].median())
    print('The standard deviation', col, 'is', df[col].std())
    print('Total no of missing values for', col, 'is', df[col].isnull().sum())
```

```python
def plot_cat(col):
    plt.figure(figsize = (20,20))
    plt.subplot(2,1,1)
    ax = sb.countplot(x = df[col])
    for bar in ax.patches:
        count = bar.get_height()
        x = bar.get_x() + bar.get_width()/2
        y = bar.get_height()
        ax.annotate(count, (x,y), va = 'bottom', ha = 'center')
    plt.title(f'Total Count of {col}');

    plt.subplot(2,1,2)
    labels = list(df[col].value_counts().index)
    plt.pie(df[col].value_counts(), labels = labels, autopct= "%0.2f%%");
    plt.title(f'Percentage of {col}');
```

```python
def plot_con(col):
    plt.figure(figsize = (20,7))
    plt.subplot(121)
    sb.histplot(x = df[col])
    plt.title(f"Histogram of {col}");

    plt.subplot(122)
    sb.boxplot(x = df[col], showfliers = False);
    plt.title(f'Box Plot of {col}');
```
