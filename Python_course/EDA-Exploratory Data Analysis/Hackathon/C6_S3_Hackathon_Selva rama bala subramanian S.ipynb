```python
def plot_cat(col):
    plt.figure(figsize = (20,7))
    plt.subplot(1,2,1)
    ax = sb.countplot(x = city_df[col])
    for bar in ax.patches:
        count = bar.get_height()
        x = bar.get_x() + bar.get_width()/2
        y = bar.get_height()
        ax.annotate(count, (x,y), va = 'bottom', ha = 'center')
    plt.title(f'Total Count of property {col} wise');

    plt.subplot(1,2,2)
    labels = list(city_df[col].value_counts().index)
    plt.pie(city_df[col].value_counts(), labels = labels, autopct= "%0.2f%%", explode = [0.2,0], shadow = True);
    plt.title(f'{col} wise property ratio');
```

### Univariate Analysis:

```python
def info_cat(col):
    print('Unique values for the', col, 'is', df[col].unique())
    print('Mode for the', col, 'is', df[col].mode()[0])
    print('Total count of missing values in', col, 'is', df[col].isnull().sum())
```

```python
def info_con(col):
    print('The mean of', col, 'is', df[col].mean())
    print('The mode of', col, 'is', df[col].mode()[0])
    print('The median of', col, 'is', df[col].median())
    print('The standard deviation', col, 'is', df[col].std())
    print('Total no of missing values for', col, 'is', df[col].isnull().sum())
```

### Analysis of continuous variables:

```python
def plot_num(col):
    plt.figure(figsize=(20,7))
    plt.subplot(121)
    ax = sb.histplot(x = col, data = city_df);
    for bar in ax.patches:                                             
        count = (bar.get_height())                                                                           
        x = bar.get_x() + bar.get_width() /2                            
        y = bar.get_height()                                           
        ax.annotate(count, (x, y), va = "bottom", ha ="center" )    
    plt.title(f"Histogram of {col}");

    plt.subplot(122)
    sb.boxplot(x = city_df[col], showfliers = False);
    plt.title(f'Box Plot of {col}');
```

```python
def cat_cat_plot(col1, col2):
    plt.figure(figsize = (20,7))
    ax = sb.countplot(x = col2, hue = col1, data = city_df)
    for bar in ax.patches:                                             
        count = (bar.get_height())                                                                           
        x = bar.get_x() + bar.get_width() /2                            
        y = bar.get_height()                                           
        ax.annotate(count, (x, y), va = "bottom", ha ="center" )    
    plt.title(f'Total count of {col2} wrt {col1}');
```

### Analysis of Animal on City:

```python
import warnings
warnings.filterwarnings('ignore')
def cat_con(col1, col2): 
    plt.figure(figsize=(20,7))
    plt.subplot(121)
    sb.histplot(x = col1, hue = col2, data = city_df, bins = 5)
    plt.title(f'Histogram of {col1} wrt {col2}');

    plt.subplot(122)
    sb.boxplot(y = city_df[col1], x = city_df[col2], showfliers = False);
    plt.title(f'Box Plot of {col1} wrt {col2}');
```
