# Data Processing

### Build year seperate by range wise :

```python
df1=df[df["Built Year"].between(1900,1925)]
year1900_1925=df1['Built Year'].count()

df2=df[df["Built Year"].between(1925,1950)]
year1925_1950=df2['Built Year'].count()

df3=df[df["Built Year"].between(1925,1975)]
year1925_1975=df3['Built Year'].count()

df4=df[df["Built Year"].between(1975,2000)]
year1975_2000=df4['Built Year'].count()

df5=df[df["Built Year"].between(2000,2005)]
year2000_2005=df5['Built Year'].count()

df6=df[df["Built Year"].between(2005,2010)]
year2005_2010=df6['Built Year'].count()

df7=df[df["Built Year"].between(2010,2015)]
year2010_2015=df7['Built Year'].count()
```

```python
year=pd.DataFrame({"Built Year":['1900-1925','1925-1950','1950-1975','1975-2000','2000-2005','2005-2010','2010-2015'],
                   "Count":[year1900_1925,year1925_1950,year1925_1975,year1975_2000,year2000_2005,year2005_2010,year2010_2015]})
plt.figure(figsize = (12,8))
sns.barplot(x = 'Built Year', y = 'Count', data = year)
plt.show()
```

### Total No.of Bathrooms percentage wise

```python
fig, ax = plt.subplots(figsize= (20, 8))
ax = sns.countplot(x = df["number of bathrooms"])
for bar in ax.patches:                                              # ax.patches returns the rectange object from matplotlib
    percentage = f"{round(bar.get_height() /len(df), 2)}%"          # first it calculate the bar height and then dividing by the number of rows in data(to calculate percentage) 
                                                                    # and multiply with 100 to get the percentage and rounding of by 2 numbers.  
    x = bar.get_x() + bar.get_width() /2                            # get_x: returns the left corrdinate of the rectange, get_width: returns width of rectagle
    y = bar.get_height()                                            # get_height: returns height of rectange
    ax.annotate(percentage, (x, y), va = "bottom", ha ="center" )   # va bottom: means percentage would be above the bar head. 
                                                                    # annotate: used to annotate over objects, here objects are the bars
plt.show()
```
